```python
from itertools import cycle, count


def tidy_rays(rs):
    series_map = {
        'X': [],
        'Y': [],
        'Z': [],
        'color': []
    }
    colors = count()
    for (origin, direction), color in zip(rs, colors):
        xo, yo, zo = origin
        xd, yd, zd = direction
        
        series_map['X'].extend((xo, xd + xo))
        series_map['Y'].extend((yo, yd + yo))
        series_map['Z'].extend((zo, zd + zo))
        
        series_map['color'].extend((color, color))
    
    return series_map
```
